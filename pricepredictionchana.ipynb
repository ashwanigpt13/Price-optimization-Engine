```python
import sklearn
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import TimeSeriesSplit
# def time_based_train_test_split(data, test_size=0.2):
#     split_index = int(len(data) * (1 - test_size))
#     train_data, test_data = data.iloc[:split_index], data.iloc[split_index:]
#     return train_data, test_data
# train_data, test_data = time_based_train_test_split(chana_prices, test_size=0.05)

# dummies_fruits = pd.get_dummies(fruit_prices[['productname']],dtype=float)
# dummies_fruits
# dummies_fruits.drop(['productname_Asparagus'],axis=1)
# fruits_data = pd.concat([fruit_prices,dummies_fruits],axis=1)
# fruits_data = fruits_data[0:400]
# fruits_train_data,fruits_test_data = time_based_train_test_split(fruits_data,test_size=0.02)
# fruits_data 
total_data_points = chana_prices.shape[0]
train_points = int(0.95*total_data_points)
number_of_train_point = min(365,train_points)
test_points = total_data_points - train_points

train_data = chana_prices.iloc[:train_points]
sarima_model = sm.tsa.SARIMAX(train_data['Prices'], order=(1, 1, 1), seasonal_order=(0, 0, 0, 0))
sarima_results = sarima_model.fit()


predicted=[]
actual=[]
errorP = []
# Predict and update model daily for the next 6 days (131st to 136th)
for i in range(train_points, total_data_points):
    # Get the actual market price of the current day
    actual_price = chana_prices.iloc[i]['Prices']
    
    # Predict the market price for the current day
    predicted_price = sarima_results.forecast(steps=1).values[0]
    
    # Print the predicted and actual prices
    print(f"Day {i + 1} - Predicted price: {predicted_price}, Actual price: {actual_price}")
    
    # Calculate the difference between predicted and actual prices
    price_difference = actual_price - predicted_price
    price_percentage_error = abs(price_difference/actual_price)
    predicted.append(predicted_price)
    actual.append(actual_price)
    errorP.append(price_percentage_error)
    # Update training data with the actual price of the current day
    train_data = pd.concat([train_data, chana_prices.iloc[[i]]], ignore_index=True)
    
    # Remove the oldest data point from training data
    train_data = train_data.iloc[1:]
    
    # Update SARIMA model with the updated training data
    sarima_model = sm.tsa.SARIMAX(train_data['Prices'], order=(1, 1, 1), seasonal_order=(0, 0, 0, 0))
    sarima_results = sarima_model.fit()

    
    # Print the difference between actual and predicted prices
    print(f"percent error between actual and predicted price: {price_percentage_error}\n")
print(predicted)
print(actual)
print(errorP)
```
